```python
import matplotlib.pyplot as plt
from datetime import datetime
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
```

```python
cg.get_price(ids='hive,steem',vs_currencies='usd,btc',include_market_cap='true')
```

```python
hivemc = cg.get_coin_market_chart_by_id(id='hive', vs_currency='usd', days=90)
steemmc = cg.get_coin_market_chart_by_id(id='steem', vs_currency='usd', days=90)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(8,6)
ax.plot([datetime.utcfromtimestamp(x[0]/1000) for x in hivemc['prices']], 
        [x[1] for x in  hivemc['prices']], label='HIVE', color='red')
ax.plot([datetime.utcfromtimestamp(x[0]/1000) for x in steemmc['prices']], 
        [x[1] for x in  steemmc['prices']], label='STEEM', color='blue')
ax.set_ylabel('Price (US$)')
ax.legend();
```

```python
fig, ax = plt.subplots()
ax.plot([(datetime.utcfromtimestamp(x/1000),y) for [x,y] in hivemc['prices']], label='HIVE')
#ax.plot([datetime.utcfromtimestamp(x/1000),y for x,y in steemmc['prices']], [x[1] for x in  steemmc['prices']], label='STEEM')
ax.set_xlabel('Hours')
ax.set_ylabel('Price')
ax.legend();
```
