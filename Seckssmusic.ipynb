```python
regex = r'\*\*[A-Z]((.|\n)*?)\*\*'
regex2 = r'##((.)*?)\n'
```

```python
with open('pw.txt') as f:
    for l in f:
        uname, pw = l.split()

cnxn_str = ("Driver={ODBC Driver 17 for SQL Server};"
            "Server=vip.hivesql.io;"
            "Database=DBHive;"
            f"UID={uname};"
            f"PWD={pw};")
cnxn = pyodbc.connect(cnxn_str)
```

```python
query="""select c.title , c.created ,
c.body , c.permlink, c.url
from
    Tags
    INNER JOIN Comments c ON Tags.comment_id = C.ID
WHERE
    Tags.tag = 'newtunes'
    and c.author = 'seckorama'
    and  c.[depth] =0
    and c.created > '2018-01-01'
order by c.created """
```
